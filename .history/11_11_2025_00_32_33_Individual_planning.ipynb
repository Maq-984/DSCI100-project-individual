### 1.0 - Data Description

The dataset includes two files:
- **players.csv** — 196 observations and 7 variables. Each row represents one unique player.  
- **sessions.csv** — 1,535 observations and 5 variables. Each row represents one gameplay session linked to a player by `hashedEmail`.

#### Variables Summary

|  | hashedEmail | character | Player's hashed Email |  |
|  | played_hours | numeric | Total hours played | inaccurate values, since one decimal |

|  | gender | character | gender | |
