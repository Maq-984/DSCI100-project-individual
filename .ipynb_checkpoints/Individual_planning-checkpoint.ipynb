### 3.1 necessary wrangling


```r
players_tidy <- players |>
    rename(
        hashed_email = hashedEmail,
        age = Age
    )

sessions_tidy <- sessions |>
    rename(
        hashed_email = hashedEmail,
    ) |> 
    mutate(
    start_time = as.POSIXct(start_time, format="%d/%m/%Y %H:%M"),
    end_time = as.POSIXct(end_time, format="%d/%m/%Y %H:%M"),
    duration_hrs = round (as.numeric(difftime(end_time, start_time, units = "hours")), 2)
  )

head(players_tidy, n = 6)
head(sessions_tidy, n = 6)
```

### 3.2 Calculating average of each quantitve value in players.csv



```r
players_avgs <- players_tidy |>
    summarise(avg_age = mean(age, na.rm = TRUE),
             avg_played_hrs = mean(played_hours, na.rm = TRUE))

players_avgs
```

### 3.3 Visualizing

#### Relation between Subscription Status & Experience Level

```r
experience_subscription_plot <- players_tidy |>
    ggplot(aes (x = experience, fill = subscribe))+
    geom_bar(position = "dodge")+
    labs(
        title = "Subscription Status by Experience Level",
        x = "Experience Level",
        y = "Number of Players",
        fill = "Subscribed"
    ) + 
    theme(
    plot.title = element_text(face = "bold", hjust = 0.5),
    axis.text.x = element_text(angle = 45, hjust = 1),
    axis.text = element_text(size = 10),
    axis.title = element_text(size = 11)
  )
experience_subscription_plot
```

Insight : Across different experience levels, number of players subscribed is more than double than those who are not subscribed.

#### Relation between Age & Played Hours

```r
age_played_hrs_plot <- players_tidy |>
    ggplot(aes(x = age, y = played_hours))+
    geom_point(alpha = 0.35, color = "blue")+
    labs(title = "Relation between Age & Played Hours",
        x = "Age (years)",
        y = "Played Hours (hrs)")+
    theme(
    plot.title = element_text(face = "bold", hjust = 0.5),
    axis.text.x = element_text(angle = 45, hjust = 1),
    axis.text = element_text(size = 10),
    axis.title = element_text(size = 11)
  )

age_played_hrs_plot
```

Insight : Younger players usually plays more hours than older players.

#### Played Hours & Experience Level

```r
played_hrs_subscription_plot <- avg_hours_exp <- players |>
  group_by(experience) |>
  summarise(mean_hours = mean(played_hours, na.rm = TRUE))

ggplot(avg_hours_exp, aes(x = experience, y = mean_hours, fill = experience)) +
  geom_bar(stat = "identity") +
  labs(
    title = "Average Played Hours by Experience Level",
    x = "Experience Level",
    y = "Average Played Hours (hours)"
  ) +
    theme(
    plot.title = element_text(face = "bold", hjust = 0.5),
    axis.text.x = element_text(angle = 45, hjust = 1),
    axis.text = element_text(size = 10),
    axis.title = element_text(size = 11)
  )
```

Insight : There is no clear trend but regular players average playing hours are much higher comparing to each other experience level.

#### Proportion of subsribed players across different playerd hours ranges

```r
ggplot(players, aes(x = played_hours, fill = subscribe)) +
  geom_histogram(position = "fill", bins = 50, color = "white") +
  labs(
    title = "Proportion of Subscribed Players by Played Hours",
    x = "Played Hours (hours)",
    y = "Proportion of Players",
    fill = "Subscribed"
  )+
    theme(
    plot.title = element_text(face = "bold", hjust = 0.5),
    axis.text.x = element_text(angle = 45, hjust = 1),
    axis.text = element_text(size = 10),
    axis.title = element_text(size = 11)
  )
```

Insight : players who spend more playing hours are likely to subscribe.

#### Portion of subscription by Age range

```r
ggplot(players_tidy, aes(x = cut(age, breaks = seq(0, 80, by = 5)), fill = subscribe)) +
  geom_bar(position = "fill", color = "white") +
  labs(
    title = "Proportion of Subscribed Players by Age Range",
    x = "Age Range (years)",
    y = "Proportion of Players",
    fill = "Subscribed"
  ) +
    theme(
    plot.title = element_text(face = "bold", hjust = 0.5),
    axis.text.x = element_text(angle = 45, hjust = 1),
    axis.text = element_text(size = 10),
    axis.title = element_text(size = 11)
  )
```

Insight : Younger players are more likely to subscribe, while subscription rates drop after 25.
