### 1.0 - Data Description

The dataset includes two files:
- **players.csv** — 196 observations and 7 variables. Each row represents one unique player.  
- **sessions.csv** — 1,535 observations and 5 variables. Each row represents one gameplay session linked to a player by `hashedEmail`.

#### Variables Summary

| File | Variable | Type | Description | Possible Issues |
|------|-----------|------|--------------|-----------------|
| players.csv | experience | character | Player’s experience level (Amateur, Regular, Pro, etc.) |  |
|  | subscribe | logical | Whether the player subscribed to the newsletter | Imbalanced TRUE/FALSE values |
|  | hashedEmail | character | Player's hashed Email |  |
|  | played_hours | double | Total hours played | inaccurate values, since one decimal |
|  | name | character | Player name | No uses for it |
|  | gender | character | gender | |
|  | Age | double | Player age (years) | inaccurate values, since one decimal |
| sessions.csv | hashedEmail | character | Player Email hashed (identifier) |  |
|  | start_time, end_time | character | Human readable timestamps for session start and end |  |
|  | original_start_time, original_end_time | numeric | original time |  |


```r
experience_count <- players |>
    count(experience)

gender_count <- players |>
    count(gender)

subscribe_count <- players |>
    count(subscribe)

avg_played_hours <- mean(players$played_hours, na.rm = TRUE)
avg_age <- mean(players$Age, na.rm = TRUE)
```
